```python
import pandas as pd
import datetime as dt
import numpy as np
```

```python
# loại bỏ các biến không cần thiét
df_booking = df_booking_raw.drop(df_booking_raw.columns[[2, 3, 4, 7, 8, 18, 19, 20, 21]], axis=1)
```

```python
df_booking
```

```python
# Thay đổi data type phù hợp
# Xử lý data dạng datime time bằng thư viện pandas, và tham số errors-'coerce' để đặt thành giá trị NaN nếu không thể chuyển về dạng datime như mong muốn
df_booking['Arrival'] = pd.to_datetime(df_booking['Arrival'], errors='coerce')
df_booking['Departure'] = pd.to_datetime(df_booking['Departure'], errors='coerce')
df_booking['Rate'] = df_booking['Rate'].astype('string')
df_booking['Booked'] = df_booking['Booked'].astype('string')
# # df_booking['Vip'] = df_booking['Vip'].astype('int')
# df_booking['Group'] = df_booking['Group'].astype('int')
df_booking['Company'] = df_booking['Company'].astype('string')
df_booking['Adt/Chl/Enf'] = df_booking['Adt/Chl/Enf'].astype('string')
```

```python
# Fill các giá trị dựa vào biến 'Folio#'
df_booking['Rate'] = df_booking.groupby('Folio#')['Rate'].transform('ffill')
df_booking['Vip'] = df_booking.groupby('Folio#')['Vip'].transform('ffill')
df_booking['Booked'] = df_booking.groupby('Folio#')['Booked'].transform('ffill')
df_booking['Adt/Chl/Enf'] = df_booking.groupby('Folio#')['Adt/Chl/Enf'].transform('ffill')
df_booking['MCFLSp'] = df_booking.groupby('Folio#')['MCFLSp'].transform('ffill')
df_booking['Company'] = df_booking.groupby('Folio#')['Company'].transform('ffill')
df_booking['Sales'] = df_booking.groupby('Folio#')['Sales'].transform('ffill')
df_booking['NAT'] = df_booking.groupby('Folio#')['NAT'].transform('ffill')
df_booking['Specials'] = df_booking.groupby('Folio#')['Specials'].transform('ffill')
df_booking['Notice'] = df_booking.groupby('Folio#')['Notice'].transform('ffill')
```

```python
df_booking.info()
```

```python
df_booking
```

```python
import re
# Hàm trích xuất phần trong ngoặc
def extract_in_parentheses(text):
    if isinstance(text, str):  # Kiểm tra nếu giá trị là chuỗi
        match = re.search(r'\((.*?)\)', text)
        if match:
            return match.group(1)
    return None

# Thay thế NaN bằng chuỗi trống
df_booking['Rate'] = df_booking['Rate'].fillna('')

# Áp dụng hàm lên cột 'Rate'
df_booking['Price_Types'] = df_booking['Rate'].apply(extract_in_parentheses)
```

```python
# Từ biến 'Adt','Chl', 'Enf' tách ra làm 3 biến mới qua các dấu '/'
df_booking[['Adt','Chl', 'Enf']] = df_booking['Adt/Chl/Enf'].str.split('/', expand=True)
```

```python
df_booking
```

```python
# Loại bỏ những hàng có giá trị NaN của biến 'Adt/Chl/Enf' và 'Booked' 
df_booking = df_booking.dropna(subset=['Adt/Chl/Enf'])
df_booking = df_booking.dropna(subset=['Booked'])
df_booking = df_booking.dropna(subset=['Notice'])
# Không loại bỏ giá trị NaN của NAT bởi vì biến 'Sts' có giá trị là 'CX' thì giá trị ở 'NAT' có thể là NaN
```

```python
df_booking.drop_duplicates(subset = None, keep='first',inplace= True)
```

```python
# Xử lý những giá trị lỗi của biến 'Adt','Chl', 'Enf' về dạng chuẩn x/y/z
df_booking['Adt/Chl/Enf']= df_booking.apply(
    lambda row: f"{int(row['Adt/Chl/Enf'][8:10])}/{int(row['Adt/Chl/Enf'][5:7])}/{int(row['Adt/Chl/Enf'][2:4])}" 
                if pd.isna(row['Enf']) and len(row['Adt/Chl/Enf']) >= 10 else row['Adt/Chl/Enf'], axis=1)
```

```python
# Lặp lại thao tác tách biến 'Adt','Chl', 'Enf' tách ra làm 3 biến mới sau khi đã xử lý lỗi
df_booking[['Adt','Chl', 'Enf']] = df_booking['Adt/Chl/Enf'].str.split('/', expand=True)
```

```python
df_booking.info()
```

```python
# Kiểm tra kết quả xử lý biến 'Adt','Chl', 'Enf' bằng cách kiểm tra NaN của giá trị 'Enf' còn tồn tại không
df_booking.loc[df_booking['Enf'].isna()]
```

```python
df_booking_nd2 = pd.read_excel(r"D:\DATA ANALYST\Booking Data Processed LCR 2022-2023.xlsx")
```

```python
df_booking_nd2.head()
```

```python
# Loại bỏ những cột trùng lặp thông tin 
df_booking_nd2 = df_booking_nd2.drop(columns= ['Rate','Adt/Chl/Enf'])
```

```python
df_booking_nd2
```

```python
# Xử lý datatype 'Book time' qua datatime để thực hiện tính toán
df_booking_nd2['Book Time'] = pd.to_datetime(df_booking_nd2['Book Time'], errors='coerce')
```

```python
# Tạo 2 biến 'Days stay' và 'Lead time' qua việc tính toán các biến dạng datetime
df_booking_nd2['Days stay'] = (df_booking_nd2['Departure'] - df_booking_nd2['Arrival']).dt.days + 1
df_booking_nd2['Lead time'] = (df_booking_nd2['Arrival']- df_booking_nd2['Book Time']).dt.days + 1
```

```python
# Khai thác các biến mới, dựa vào biến ' Note'
df_booking_nd2['FREE CXL'] = df_booking_nd2['Notice'].str.contains('FREE CXL')
df_booking_nd2['TACC'] = df_booking_nd2['Notice'].str.contains('TACC')
df_booking_nd2['KTT'] = df_booking_nd2['Notice'].str.contains('KTT')
```

```python
df_booking_nd2
```

```python
# Kiểm tra biến phụ thuộc 'Sts' 
df_booking_nd2['Sts'].unique()
```

```python
# Thay thế giá trị 'NS' thành 'CX', vì có ý nghĩa tương tự nhau
df_booking_nd2['Sts'] = np.where(df_booking_nd2['Sts'].str.contains('NS'), 'CX', df_booking_nd2['Sts'])
```
